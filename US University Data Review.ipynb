
## Open data
When I heard about the open data that's been made available for US universities, I was excited.  I liked the idea of plumbing the data to find schools that provide better value than others.  I doubt that all bachelor's programs are equivalent in terms of what you can learn.  But in my opinion -- beyond your first job (and maybe even at your first job), the different schools won't distinguish you as much as your work experience does.

Let's see if we can find a way to evaluate the schools that will reveal some of the ones who do a better job than others.  As it stands, I think students have a really tough time evaluating schools critically.

### Analysis
First I started by filtering out lots of schools: ones that didn't publish all of the fields we're looking to compare, ones that aren't accredited, and ones that are "less traditional" (> 50% undergrads over 25, predominant degree not bachelor's, e.g.).  I'm sure that I may have discarded some worthwhile schools by doing this but it's a good approximation.



```python

%matplotlib inline

from __future__ import print_function
import json
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def only_interesting(school):
    BACHELOR = 3
    return school['CURROPER'] == 1 and school['DISTANCEONLY'] == 0 \
       and school['AccredAgency'] != None and len(school['AccredAgency']) > 0 \
       and school['HIGHDEG'] >= BACHELOR  and school['PREDDEG'] == BACHELOR \
       and school['AVGFACSAL'] != None and school['INEXPFTE'] != None \
       and school['SAT_AVG'] != None  and school['CCSIZSET'] >= 6 \
       and school['UG25abv'] <= 0.5 and school['TUITFTE'] != None \
       and school['GRAD_DEBT_MDN'] != None

DEBUG_SMALL_SUBSET = True

if DEBUG_SMALL_SUBSET:
    school_list = json.loads(open('out.json').read())
else:
    schools_years = json.loads(open('college_scorecard.json').read())
#   school_list = schools_years['MERGED2001_PP.csv']
    school_list = schools_years['MERGED2012_PP.csv']

schools = {school['INSTNM']: school for school in school_list if only_interesting(school) }

school_names = sorted(schools.keys())
print(len(school_list), 'reduced to', len(school_names))


def by_inexpfte_over_tuit(school):
    return float(school['INEXPFTE']) / float(school['TUITFTE'])

def by_sat_over_tuit(school):
    return school['SAT_AVG'] / float(school['TUITFTE'])

def by_adm_rate_over_tuit(school):
    return school['ADM_RATE'] / float(school['TUITFTE'])

def by_fac_over_cost(school):
    return school['AVGFACSAL'] / (school['COSTT4_P'] if school['COSTT4_P'] else school['COSTT4_A']) 

def by_fac_over_tuit(school):
    '''Returns per-student average faculty salary over tuition-per-student
    '''
#    fac_sal_per_student = float(school['AVGFACSAL'])/ school['UGDS']
#    return fac_sal_per_student / float(school['TUITFTE'])
    return float(school['AVGFACSAL']) / float(school['TUITFTE'])

def by_fac_over_inexpfte(school):
    '''Returns average faculty salary over instructional expenses per student
    '''
    return school['AVGFACSAL'] / float(school['INEXPFTE'])

def clever_index(school):
    '''Ultimately an attempt to define a school's utility to its
       customer.  We'll try to minimize debt, maximize per-tuition
       instructional expenditures, maximize average faculty salary
       per school.
    '''
    # weighted sum of various parameters
    return (.3 / school['GRAD_DEBT_MDN']) \
        + (by_inexpfte_over_tuit(school) * 0.4) \
        + (by_fac_over_tuit(school) * 0.3)
        
def by_tuit(school):
    return school['TUITFTE']

def by_debt(school):
    return school['GRAD_DEBT_MDN']

```

## Assumptions
The fields used, and some assumptions regarding them:
 * Tuition is 'TUITFTE' which is actual mean tuition paid per full-time equivalent student.  We can't decompose in-state versus out-of-state rates from this data.
 * Median debt is 'GRAD_DEBT_MDN', describing the debt held by students who completed the program.
 * Undergraduate student body is 'UG25abv' describing the ratio of students who are older than 25.
    
Those fields make sense for my focus, which is to understand which schools will give the greatest value.  The school's value is likely a function of its tutilage capability and a student's opportunity to network with their colleagues and a school's alumni.  Unfortunately we don't have any data regarding most of this, so we'll make very, *very* gross generalizations.

### To what end?
Well, by exploring this data, we can offer insight into schools that a candidate might not otherwise consider. 

```python
inexpfte_over_tuit = [by_inexpfte_over_tuit(schools[school_name]) for school_name in school_names]
fac_over_tuit = [by_fac_over_tuit(schools[school_name]) for school_name in school_names]
fac_over_inexpfte = [by_fac_over_inexpfte(schools[school_name]) for school_name in school_names]
adm_rate = [schools[name]['ADM_RATE'] for name in school_names]
sat_avg = [schools[name]['SAT_AVG'] for name in school_names]
share_older = [schools[name]['UG25abv'] for name in school_names]
debt = [by_debt(schools[school_name]) for school_name in school_names]
tuit = [by_tuit(schools[school_name]) for school_name in school_names]
inexpfte = [schools[school_name]['INEXPFTE'] for school_name in school_names]
earnings = [schools[school_name]['md_earn_wne_p10'] for school_name in school_names]
ugds = [schools[school_name]['UGDS'] for school_name in school_names]
score = [clever_index(schools[school_name]) for school_name in school_names]

#print([x for x in dir((sns)) if 'plot' in x])

#sns.distplot(tuit_over_inexpfte)
#sns.distplot(fac_over_tuit)

#plt.scatter(x=range(len(fac_over_tuit)), y=fac_over_tuit, color='blue')
#plt.scatter(x=range(len(tuit_over_inexpfte)), y=tuit_over_inexpfte, color='green')
#plt.twinx()
#plt.scatter(x=range(len(debt)), y=tuit_over_inexpfte, color='aqua')

#plt.plot(tuit_over_inexpfte, color='blue')
#plt.twinx()
#plt.plot(debt, color='red')
#plt.plot(tuit, color='green')
school_general = pd.DataFrame(data={'facsalary/tuit': fac_over_tuit,
                                    'facsalary/inexp': fac_over_inexpfte,
                                    'inexp/tuit': inexpfte_over_tuit,
                                    'debt_usd': debt,
                                    'tuit_usd': tuit,
                                    'inexpfte_usd': inexpfte,
                                    'adm_rate': adm_rate,
                                    'share_older': share_older,
                                    'sat_avg': sat_avg,
                                    'name': school_names,
                                    'student_count': ugds,
                                    'med_earn': earnings,
                                    'score': score,
                                    })
#school_excluding_top_extreme_names = [school_name for school_name in school_names if schools[school_name]['SAT_AVG'] < 1250]

school_excluding_top_extreme = school_general.query('sat_avg < 1250')

school_factors = pd.DataFrame(data={'facsalary/tuit': fac_over_tuit,
                                    'facsalary/inexp': fac_over_inexpfte,
                                    'inexp/tuit': inexpfte_over_tuit,
                                    'name': school_names,
                                    })
school_dollars = pd.DataFrame(data={
                                    'debt_usd': debt,
                                    'tuit_usd': tuit,
                                    'name': school_names,
                                    })
                            

```

## Value

Let's examine which schools provide better value than others.

One way we could consider value is how much of your tuition is devoted to your tutilage.

Let's consider
$$Value = \frac{instructional\ expenses_{FTE}}{Tuition_{FTE}}$$

Interestingly, there are schools on both sides of 1:1.  There are some schools whose instructional expenses are quite a bit more than their tuition.  There's public schools, it makes some sense that they are only partially funded by tuition.  Public funding for education is actually a positive externality.

Interestingly enough, there's private schools on this list too.  Cal Tech, Yale -- they are schools with big endowments who may use that to offset some tuition cost.

```python
factors_graph = sns.violinplot(school_factors)
factors_graph.set_title('School Factors')

print('Worst by instructional expenditures over tuition')
school_general.sort_values(by='inexp/tuit')[:5]
```

```python
print('Best by instructional expenditures over tuition')
school_general.sort_values(by='inexp/tuit')[-5:]
```
